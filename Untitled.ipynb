```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img, title="Image"):
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def preprocessing_image(img):
    img = cv2.imread("balloons_noisy.png")
    if img is None:
        print("Error: Image not found. Check the file path!")
        return None
    print("✅ Image loaded successfully")

    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    show(rgb, "Original Image")

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    show(gray, "Grayscale")

    denoised = gray.copy()
    for i in range(5):   
        denoised = cv2.medianBlur(denoised, 5)
    show(denoised, "After Median Filter")

    denoised = cv2.GaussianBlur(denoised, (3, 3), 0)
    show(denoised, "After Gaussian")

    final = cv2.cvtColor(denoised, cv2.COLOR_GRAY2RGB)
    show(final, "Final Image")
    return final
```

```python
img = cv2.imread("balloons_noisy.png")
final_image = preprocessing_image(img)
```
